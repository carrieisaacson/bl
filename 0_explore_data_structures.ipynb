```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('data/bl_fulfillment_orders.csv')
orders['created_at'] = pd.to_datetime(orders['created_at'])

items = pd.read_csv('data/bl_fulfillment_order_items.csv')
items['created_at'] = pd.to_datetime(items['created_at'])

states = pd.read_csv('data/bl_fulfillment_order_states.csv')
states['created_at'] = pd.to_datetime(states['created_at'])
```

### Items Table

```python
# Baby Boxes, about 15k +/- each month
items[np.logical_and(items.price == 0, items.name.str.contains('Baby Box'))][['id', 'name']]\
    .groupby('name')\
    .agg(['count'])
```

```python
## There's a handful of 0-price ordered items back in 2018. Freebies?
items[np.logical_and(items.price == 0, np.logical_not(items.name.str.contains('Baby Box'))) ]
```

```python
# Most popular paid items (exclude Baby Boxes)...Lol Sophie Giraffe!
items[np.logical_and(items.price > 0, np.logical_not(items.name.str.contains('Baby Box'))) ][['name', 'id']]\
    .groupby('name')\
    .agg('count')\
    .sort_values('id', ascending=False).head(20)
```

### States Table


Primary order states for orders w/ physical items: 
* open_order -> order started
* ordered -> baby boxes & drop ships
* prepare_shipment
    * in between prepare_shipment & shipped there can be misc operator holds add/removes or payment holds add/removes
* shipped -> order went out the door
* cancelled -> process was terminated 

Orders for cash funds & gift cards:
* completed

```python
## What order types are associated with 'ordered' states
order_ids = states.loc[states.state == 'ordered']['order_id']
orders.loc[orders.id.isin(order_ids)][['id', 'order_type']]\
    .groupby('order_type')\
    .agg(['count'])
```

```python
## What are 'completed' states --> looks like Cash & Electronic Gift Cards will go to 'Completed' status
order_ids = states.loc[states.state == 'completed']['order_id']
orders.loc[orders.id.isin(order_ids)][['id', 'order_type']]\
    .groupby('order_type')\
    .agg(['count'])
```

```python
orders.groupby('order_type')\
    .agg(['count'])
```

```python
order_ids.sample(10)
```

```python
orders[orders.id.isin(order_ids.sample(10))]
```

```python
orders[orders.id.isin(order_ids.sample(10))].sort_values(['id', 'created_at'])
```

```python
items[items.order_id.isin(order_ids.sample(10))]
```

```python
## What are 'partially shipped' states --> looks like Cash & Electronic Gift Cards will go to 'Completed' status
order_ids = states.loc[states.state == 'partially_shipped']['order_id']
orders.loc[orders.id.isin(order_ids)][['id', 'order_type']]\
    .groupby('order_type')\
    .agg(['count'])
```

```python
states[states.order_id.isin(order_ids.sample(10))].sort_values(['order_id', 'created_at'])
```
